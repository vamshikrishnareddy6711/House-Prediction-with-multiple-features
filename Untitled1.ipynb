Maximum Permutation After Removing Vowels

Interpretation used: remove vowels from the input string, then return the characters permuted to the lexicographically largest order (i.e., sort remaining characters descending).
(If you meant something else, tell me and I’ll adapt.)

```python
def max_permutation_after_removing_vowels(s: str) -> str:
    """
    Remove vowels from the string s, then return the lexicographically largest
    permutation of the remaining characters (i.e., characters sorted descending).
    Non-letter characters are kept (but vowels among letters are removed).
    """
    vowels = set("aeiouAEIOU")
    # Keep characters that are NOT vowels
    kept = [ch for ch in s if ch not in vowels]
    # For maximum lexicographic permutation, sort descending
    kept.sort(reverse=True)
    return "".join(kept)

# Example
print(max_permutation_after_removing_vowels("hello world"))  # -> "wrrllodh" (letters and space ordering may vary)
print(max_permutation_after_removing_vowels("abcdEIOUxyz"))  # -> "zyxdcb" (vowels removed, rest sorted descending)
```

Count Special Subarrays of Size 3 (a + c = b)

Count subarrays of length 3 where first + third equals the middle value.

```python
def count_special_subarrays_size_3(arr):
    """
    Count contiguous subarrays of size 3 such that a + c == b,
    where the subarray is [a, b, c].
    """
    n = len(arr)
    count = 0
    # Slide a window of size 3
    for i in range(n - 2):
        a, b, c = arr[i], arr[i+1], arr[i+2]
        if a + c == b:
            count += 1
    return count

# Example
print(count_special_subarrays_size_3([1, 3, 2, 5, 3, 8]))  # Explanation: windows [1,3,2] -> 1+2==3 => yes; [3,2,5] -> 3+5!=2, etc.
```

Maximum Number of People Using Lift Without Exceeding Capacity

Interpretation used: given a list of people weights and a lift capacity, choose the maximum number of people whose total weight does not exceed capacity. Greedy strategy: take lightest people first.

```python
def max_people_under_capacity(weights, capacity):
    """
    Given a list of weights and a capacity, return the maximum number
    of people that can fit without the total weight exceeding capacity.
    Greedy: sort ascending and add until you exceed capacity.
    """
    weights_sorted = sorted(weights)
    total = 0
    count = 0
    for w in weights_sorted:
        if total + w <= capacity:
            total += w
            count += 1
        else:
            break
    return count

# Example
weights = [60, 80, 40, 55, 90]
capacity = 200
print(max_people_under_capacity(weights, capacity))  # -> 3 (40 + 55 + 60 = 155, adding next smallest 80 would exceed)
```

Longest Substring Containing Only Consonants

Return the longest contiguous substring made only of consonants (letters that are not vowels). Non-letter characters break substrings.

```python
def longest_consonant_substring(s):
    """
    Return (longest_substring, length) for the longest contiguous substring
    that contains only consonant letters. Non-alpha characters break the substring.
    """
    vowels = set("aeiouAEIOU")
    best = ""      # best substring found
    current = []   # list of characters for current consonant run

    for ch in s:
        # If character is alphabetic and not a vowel -> consonant
        if ch.isalpha() and ch not in vowels:
            current.append(ch)
        else:
            # When sequence breaks, check if current is best
            if len(current) > len(best):
                best = "".join(current)
            current = []  # reset for the next run

    # Final check at end of string
    if len(current) > len(best):
        best = "".join(current)

    return best, len(best)

# Examples
print(longest_consonant_substring("abcde!fghij"))   # -> ('bcdfghj' or actual consonants run depending on input)
print(longest_consonant_substring("strength has e in it"))  # demonstrates non-letter and vowels splitting
```

Check if Sum of Even-Indexed Elements > Sum of Odd-Indexed Elements

Indices are 0-based (index 0 is even).

```python
def is_even_index_sum_greater(arr):
    """
    Return True if sum of elements at even indices (0,2,4,...) is greater than
    sum of elements at odd indices (1,3,5,...). Otherwise return False.
    """
    sum_even = 0
    sum_odd = 0
    for i, val in enumerate(arr):
        if i % 2 == 0:
            sum_even += val
        else:
            sum_odd += val
    return sum_even > sum_odd

# Examples
print(is_even_index_sum_greater([1, 2, 3, 1]))  # True: even indices sum = 1+3=4, odd = 2+1=3
print(is_even_index_sum_greater([5, 10, 1]))    # False: even = 5+1=6, odd = 10
```

Count Strings With More Vowels Than Consonants

Count how many strings in a list have strictly more vowels than consonants. Only alphabetic characters count; ignore other characters.

```python
def count_strings_more_vowels(strings):
    """
    Given a list of strings, return the count of strings that contain strictly
    more vowels than consonants. Only letters are considered.
    """
    vowels = set("aeiouAEIOU")
    result = 0

    for s in strings:
        vowel_count = 0
        consonant_count = 0
        for ch in s:
            if ch.isalpha():
                if ch in vowels:
                    vowel_count += 1
                else:
                    consonant_count += 1
        if vowel_count > consonant_count:
            result += 1

    return result

# Example
words = ["idea", "sky", "queue", "rhythm", "a!e?i"]
print(count_strings_more_vowels(words))  # -> 'idea'(2v,2c no), 'queue'(4v,1c yes), 'a!e?i'(3v,0c yes) => returns 2
```

Maximum Difference Between Digit Sums in an Array

Interpretation used: For an array of integers, compute each integer's digit sum (sum of decimal digits). Return the maximum difference between any two digit-sums (i.e., max_sum - min_sum).

```python
def digit_sum(n):
    """Return sum of digits of integer n (works with negative numbers too)."""
    n = abs(n)
    s = 0
    while n:
        s += n % 10
        n //= 10
    return s

def max_digit_sum_difference(arr):
    """
    Compute digit-sum for each element and return maximum difference
    between any two digit-sums. If array has fewer than 2 elements, difference is 0.
    """
    if len(arr) < 2:
        return 0
    sums = [digit_sum(x) for x in arr]
    return max(sums) - min(sums)

# Examples
print(max_digit_sum_difference([123, 45, 9, 1002]))  # digit sums: [6,9,9,3] => max 9 - min 3 = 6
print(max_digit_sum_difference([5]))                 # -> 0 (only one element)
```

Maximum ASCII Difference Between Adjacent Characters

Given a string, find the maximum absolute difference between the ASCII codes of adjacent characters.

```python
def max_ascii_diff_adjacent(s):
    """
    Return the maximum absolute ASCII difference between adjacent characters
    in string s. Also returns the index where it occurs and the pair of chars.
    If string length < 2, returns (0, None, None).
    """
    if len(s) < 2:
        return 0, None, None

    max_diff = 0
    max_index = None  # index of first character in the pair
    pair = None

    for i in range(len(s) - 1):
        diff = abs(ord(s[i]) - ord(s[i+1]))
        if diff > max_diff:
            max_diff = diff
            max_index = i
            pair = (s[i], s[i+1])

    return max_diff, max_index, pair

# Examples
print(max_ascii_diff_adjacent("azAZ!"))  # returns (difference, index, pair)
print(max_ascii_diff_adjacent("hello"))
```

1) Count Strings That Become Palindrome After Removing One Character

Interpretation / assumption: I treat this as: is there some single-character removal (exactly one) that makes the string a palindrome? (This is the common “valid palindrome II” style check; it also returns True for strings that are already palindromes because removing one character from a palindrome can still produce a palindrome.)

```python
def becomes_palindrome_after_one_removal(s: str) -> bool:
    """
    Return True if by removing exactly one character from s we can obtain a palindrome.
    We use two pointers and allow skipping one character (left or right) at the first mismatch.
    """
    def is_palindrome_range(st, left, right):
        # check if st[left:right+1] is palindrome
        while left < right:
            if st[left] != st[right]:
                return False
            left += 1
            right -= 1
        return True

    left, right = 0, len(s) - 1
    while left < right:
        if s[left] == s[right]:
            left += 1
            right -= 1
        else:
            # Try removing the left character or the right character
            return is_palindrome_range(s, left+1, right) or is_palindrome_range(s, left, right-1)
    # If we never found a mismatch, removing any one character (if length>0) yields a palindrome.
    # For empty string, there's no char to remove; return False.
    return len(s) > 0

# Examples
print(becomes_palindrome_after_one_removal("abca"))  # True: remove 'c' -> "aba"
print(becomes_palindrome_after_one_removal("racecar"))  # True: already palindrome -> remove one still palindrome
print(becomes_palindrome_after_one_removal("abc"))  # False
print(becomes_palindrome_after_one_removal(""))  # False (no char to remove)
```

) Sort Strings by Number of Distinct Characters

Assumption: Stable sort ascending by count of distinct characters; if tie, sort lexicographically for determinism.

```python
def sort_by_num_distinct(strings):
    """
    Sort strings by number of distinct characters (ascending).
    Tie-breaker: lexicographical order.
    """
    return sorted(strings, key=lambda s: (len(set(s)), s))

# Example
words = ["aaa", "abc", "de", "bba", "a", "cab"]
print(sort_by_num_distinct(words))
# Output explanation: 'a' (1 distinct), 'aaa' (1 distinct but 'a' < 'aaa'), 'de' (2), 'bba' (2 distinct 'b','a'), 'cab' (3), 'abc' (3 but lexicographically)
```

3) Longest Balanced Substring of 0s and 1s

Interpretation: longest contiguous substring with equal number of '0' and '1'. (Classic prefix-sum/hashmap trick.)

```python
def longest_balanced_01_substring(s):
    """
    Return (substring, length) for the longest contiguous substring with equal 0s and 1s.
    Approach:
      - Treat '0' as -1 and '1' as +1, compute prefix sums.
      - If the same prefix sum appears at indices i and j, the subarray (i+1..j) sums to 0 -> equal zeros and ones.
    Time: O(n)
    """
    # map prefix_sum -> first index where it occurred
    first_index = {0: -1}  # prefix sum 0 at index -1 (before string starts)
    prefix = 0
    max_len = 0
    best_range = (0, -1)

    for i, ch in enumerate(s):
        if ch == '0':
            prefix -= 1
        elif ch == '1':
            prefix += 1
        else:
            # ignore characters other than '0' and '1'; reset because substring must be contiguous 0/1 only
            first_index = {0: i}
            prefix = 0
            continue

        if prefix in first_index:
            length = i - first_index[prefix]
            if length > max_len:
                max_len = length
                best_range = (first_index[prefix] + 1, i)
        else:
            first_index[prefix] = i

    if max_len == 0:
        return "", 0
    l, r = best_range
    return s[l:r+1], max_len

# Examples
print(longest_balanced_01_substring("110100110"))  # may return ("1010", 4) or another correct longest balanced substring
print(longest_balanced_01_substring("000111"))    # ("000111", 6)
print(longest_balanced_01_substring("0101011"))   # ("010101", 6)
```

Check if Any Two Numbers Sum Up to K

Classic two-sum existence check. Time O(n) with set.

```python
def any_two_sum_to_k(nums, k):
    """
    Return True if there exist two distinct elements in nums whose sum equals k.
    Use a set to store seen numbers and check complement (k - current).
    """
    seen = set()
    for x in nums:
        if k - x in seen:
            return True
        seen.add(x)
    return False

# Examples
print(any_two_sum_to_k([2, 7, 11, 15], 9))  # True (2+7)
print(any_two_sum_to_k([1, 4, 5], 10))      # False
```

Remove Duplicate Characters to Form Lexicographically Smallest String

Interpretation: return the lexicographically smallest string that contains each character exactly once and preserves relative order relative to the chosen characters (classic "smallest subsequence of distinct chars" / monotonic stack problem).

```python
def remove_duplicate_letters_smallest(s):
    """
    Return the lexicographically smallest result containing every distinct char exactly once.
    Approach:
      - Keep a stack (result) and a set of included chars.
      - For each char, if not included, pop from stack while top > current char and top appears later again (so safe to remove).
      - Then push current char and mark included.
    """
    last_index = {ch: i for i, ch in enumerate(s)}  # last occurrence of each char
    stack = []
    included = set()

    for i, ch in enumerate(s):
        if ch in included:
            continue
        # try to remove larger characters from stack if they appear later again
        while stack and stack[-1] > ch and last_index[stack[-1]] > i:
            removed = stack.pop()
            included.remove(removed)
        stack.append(ch)
        included.add(ch)

    return "".join(stack)

# Examples
print(remove_duplicate_letters_smallest("cbacdcbc"))  # -> "acdb"
print(remove_duplicate_letters_smallest("bcabc"))     # -> "abc"
```

Count Pairs Whose XOR is Even

Observation: XOR of two integers is even iff their least significant bits are equal ⇨ both even or both odd. So count pairs among evens and among odds.

```python
def count_pairs_xor_even(nums):
    """
    Return count of unordered pairs (i<j) such that nums[i] XOR nums[j] is even.
    Count pairs where both numbers are even plus pairs where both are odd.
    """
    evens = sum(1 for x in nums if x % 2 == 0)
    odds = len(nums) - evens

    def nC2(n):
        return n * (n - 1) // 2

    return nC2(evens) + nC2(odds)

# Examples
print(count_pairs_xor_even([1,2,3,4,5]))  # evens={2,4}->1 pair, odds={1,3,5}->3 pairs => total 4
print(count_pairs_xor_even([2,2,2]))      # 3 pairs (all even)
```

Maximum Product of Two Distinct Elements in Array

Interpretation: choose two different indices (i != j) to maximize product. Works with negative numbers as well.

```python
def max_product_two(nums):
    """
    Return maximum product of two distinct elements (by index) from nums.
    Consider both:
      - product of two largest numbers
      - product of two smallest numbers (negative * negative)
    Handle arrays with <2 elements by returning None.
    """
    if len(nums) < 2:
        return None

    # initialize
    max1 = max2 = float('-inf')  # largest and second largest
    min1 = min2 = float('inf')   # smallest and second smallest

    for x in nums:
        # update max1, max2
        if x > max1:
            max2 = max1
            max1 = x
        elif x > max2:
            max2 = x

        # update min1, min2
        if x < min1:
            min2 = min1
            min1 = x
        elif x < min2:
            min2 = x

    return max(max1 * max2, min1 * min2)

# Examples
print(max_product_two([5, 20, 2, 6]))     # 20*6 = 120
print(max_product_two([-10, -20, 5, 3]))  # (-10)*(-20)=200
print(max_product_two([7]))               # None
```

Validate Parentheses String With One Allowed Mismatch

Assumption: The string contains only '(' and ')'. We return True if the string is already a valid parentheses string or if you can remove exactly one character (one allowed mismatch) and make it valid. (This is a simple and common interpretation.)

Approach: First check if already valid. If not, try removing each character once and check validity. This is simple and clear (O(n²) worst-case) and easy for beginners to understand.

```python
def is_valid_parentheses(s: str) -> bool:
    """Return True if s is a valid parentheses string (classic stack/balance check)."""
    balance = 0
    for ch in s:
        if ch == '(':
            balance += 1
        elif ch == ')':
            balance -= 1
        else:
            # if other chars appear, treat as invalid in this problem
            return False
        if balance < 0:  # more ) than ( at some point
            return False
    return balance == 0

def valid_with_one_removal(s: str) -> bool:
    """
    Return True if s is valid parentheses or can become valid by removing exactly one character.
    Brute force: try removing each index once and test validity.
    Complexity: O(n^2) in worst case, but simple and reliable.
    """
    if is_valid_parentheses(s):
        return True
    # try removing each single character
    for i in range(len(s)):
        t = s[:i] + s[i+1:]
        if is_valid_parentheses(t):
            return True
    return False

# Examples
print(valid_with_one_removal("()()"))    # True (already valid)
print(valid_with_one_removal("())()"))   # True (remove the extra ')')
print(valid_with_one_removal("((("))     # False (removing one won't balance)
print(valid_with_one_removal(")"))       # False (removing the one char yields empty string which is valid -> if you want to treat that as True replace logic accordingly)
```

Maximum Frequency Difference in a String

Interpretation: Compute the difference between the most frequent character count and the least frequent character count among characters that appear (i.e., max_count - min_count).

```python
from collections import Counter

def max_frequency_difference(s: str) -> int:
    """
    Return difference between highest and lowest frequency among characters that appear.
    If string is empty, return 0.
    """
    if not s:
        return 0
    freq = Counter(s)
    counts = freq.values()
    return max(counts) - min(counts)

# Examples
print(max_frequency_difference("aabbccc"))  # 3 - 1 = 2
print(max_frequency_difference("aaaa"))     # 4 - 4 = 0
print(max_frequency_difference(""))         # 0
```

Count Numbers With Exactly K Set Bits

Assumption: Input list contains non-negative integers. Count numbers whose binary representation has exactly k 1-bits (set bits).

```python
def count_with_k_set_bits(nums, k):
    """
    Count numbers in nums that have exactly k set bits in their binary representation.
    Works for non-negative integers.
    """
    count = 0
    for x in nums:
        # bin(x) returns string like '0b101', so count '1'
        if bin(x).count('1') == k:
            count += 1
    return count

# Examples
print(count_with_k_set_bits([0,1,2,3,4,7,8], 2))  # numbers with exactly 2 ones: 3(11), 5? not present, so returns 1 here for 3 only
print(count_with_k_set_bits([3,5,6,7], 2))       # 3(11),5(101),6(110) -> 3 numbers
```

First Non-Repeating Character in Entire String Array

Interpretation: Given an array/list of strings, consider the characters in the order they appear across the array (left-to-right, string-by-string). Return the first character (the earliest by position) that occurs exactly once across the entire concatenated text. If none exists, return None.

```python
from collections import Counter

def first_non_repeating_char_in_array(strings):
    """
    Return the first character (by concatenated order) that appears exactly once overall.
    If no such character, return None.
    """
    # Build full concatenated string to get global counts
    full = "".join(strings)
    if not full:
        return None
    freq = Counter(full)
    # Find first char in concatenation with count == 1
    for ch in full:
        if freq[ch] == 1:
            return ch
    return None

# Examples
print(first_non_repeating_char_in_array(["abc", "de", "b"]))  # 'a' (order: a b c d e b -> counts: a:1, b:2, c:1, d:1, e:1; 'a' is first unique)
print(first_non_repeating_char_in_array(["aa", "bb"]))       # None
```

Count numbers whose reversed value is also present in the array

Interpretation: For each non-negative integer x in the array, compute rev(x) (decimal digits reversed, e.g. 120 -> 21) and count how many x have rev(x) also present in the array. Each x is counted independently (if both x and rev(x) appear multiple times, each occurrence of x counts).

```python
def reverse_number(n: int) -> int:
    """Reverse decimal digits of a non-negative integer n."""
    rev = 0
    while n > 0:
        rev = rev * 10 + (n % 10)
        n //= 10
    return rev

def count_with_reversed_present(nums):
    """
    Return count of numbers whose reversed decimal value is also present in nums.
    e.g., for nums = [12, 21, 13], result = 2 (12 has 21 present, 21 has 12 present)
    """
    s = set(nums)
    count = 0
    for x in nums:
        if reverse_number(x) in s:
            count += 1
    return count

# Examples
print(count_with_reversed_present([12, 21, 120, 21, 3, 30]))  # 12->21 yes, 21->12 yes, 120->021->21? reverse(120)=21 so yes, etc.
# Note: leading zeros are dropped by reversing (120 -> 21).
```

Find the longest prefix where sum of digits is strictly increasing

Interpretation: Input is a list of non-negative integers. For each element compute the sum of its decimal digits. Find the length of the longest prefix (starting at index 0) such that these digit-sums form a strictly increasing sequence. Return that prefix length (or the prefix itself if desired; here I return both).

```python
def digit_sum(n: int) -> int:
    """Return sum of decimal digits of non-negative integer n."""
    s = 0
    while n:
        s += n % 10
        n //= 10
    return s

def longest_prefix_strictly_increasing_digit_sums(nums):
    """
    Return (prefix_list, length) for the longest prefix starting at index 0
    such that the digit sums are strictly increasing.
    """
    if not nums:
        return [], 0
    prefix = [nums[0]]
    prev_sum = digit_sum(nums[0])
    for x in nums[1:]:
        s = digit_sum(x)
        if s > prev_sum:
            prefix.append(x)
            prev_sum = s
        else:
            break
    return prefix, len(prefix)

# Examples
print(longest_prefix_strictly_increasing_digit_sums([5, 16, 28, 119, 34]))
# digit sums: [5,7,10,11,7] -> longest prefix with strictly increasing sums is [5,16,28,119] length 4
```

Count strings where consonant count is a prime number

Interpretation: For each string, count alphabetic consonants (letters that are letters and not vowels). Count how many strings have the consonant count equal to a prime number. (We treat a,e,i,o,u case-insensitively as vowels; other letters are consonants.)

```python
def is_prime(n: int) -> bool:
    """Simple primality test for n >= 0. 0 and 1 are not prime."""
    if n < 2:
        return False
    if n in (2, 3):
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True

def count_strings_with_prime_consonants(strings):
    """
    Count how many strings have a prime number of consonants.
    Only alphabetic characters are considered; other chars ignored.
    """
    vowels = set("aeiouAEIOU")
    result = 0
    for s in strings:
        consonant_count = 0
        for ch in s:
            if ch.isalpha() and ch not in vowels:
                consonant_count += 1
        if is_prime(consonant_count):
            result += 1
    return result

# Examples
print(count_strings_with_prime_consonants(["abc", "aei", "hello", "rhythm"]))
# "abc": consonants=2 -> prime(2)=True
# "aei": consonants=0 -> not prime
# "hello": consonants=3 -> prime
# "rhythm": consonants=6 -> not prime
# => returns 2
```

Check if rearranging characters of a string can form a palindrome

Classic rule: A string can be permuted to form a palindrome if at most one character has an odd count (for even-length strings none odd; for odd-length one odd allowed).

```python
from collections import Counter

def can_rearrange_to_palindrome(s: str) -> bool:
    """
    Return True if some permutation of s is a palindrome.
    Only character counts matter; this works for any characters.
    """
    freq = Counter(s)
    odd_counts = sum(1 for cnt in freq.values() if cnt % 2 == 1)
    # At most one odd count allowed
    return odd_counts <= 1

# Examples
print(can_rearrange_to_palindrome("carrace"))  # True ("racecar")
print(can_rearrange_to_palindrome("abc"))      # False
```
